2.   変数`size`の値を変えることで自由に図形の大きさを変更できること




### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
size = 5

for i in range(size + 1):
    print("*" * i)
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
size = 5

for i in range(size):
    print(" " * i + "*")
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 5

for i in range(size):
    space = (size - 1) - i
    print(" " * space + "*")
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 5

for i in range(size):
    row = [" "] * size # 各文字列を格納するためのリスト
    
    row[i] = "*" # 左上から右下の線(行の番号と列の番号が同じ座標)
    
    row[size - 1 - i] = "*" # 右上から左下の線(列番号がsize - 1 -i の座標)
    
    # 2つのリストを結合
    print("".join(row))
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
size = 5

for i in range(size):
    space = " " * (size - 1 - i) # 空白の数
    star = "*" * (2 * i + 1) # "*"の数(奇数個になるように計算)
    
    print(space + star)
````
